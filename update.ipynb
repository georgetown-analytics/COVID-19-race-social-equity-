```python
#Change some series data types to datetime
df['onset_dt'] = pd.to_datetime(df['onset_dt'], yearfirst = True)
df['pos_spec_dt'] = pd.to_datetime(df['pos_spec_dt'], yearfirst = True)
df['cdc_report_dt'] = pd.to_datetime(df['cdc_report_dt'], yearfirst = True, utc = True)
df.info()
```

```python
#it looks like most of the descriptive data (columns hosp_yn to medcond_yn is missing)
for x in range(6,27):
    print(df.iloc[:,x].value_counts())
```

```python
pd.options.display.float_format = '{:.2f}'.format
pd_pivot = pd.pivot_table(df, columns=['race_ethnicity_combined'], index=['res_state'], values ='cdc_report_dt', aggfunc = 'count')
pd_pivot['PercentUnknown'] = pd_pivot['Unknown']/pd_pivot.sum(axis=1)
pd_pivot['TotalCases'] = pd_pivot.sum(axis=1)
pd_pivot.sort_values(by=['TotalCases'], ascending = False)
```

```python
df_lab_confirmed = df.loc[df['current_status'] == 'Laboratory-confirmed case']
df_lab_confirmed['death_yn'].value_counts()
```

```python
#confirmed cases by sex
df_lab_confirmed = df[df['current_status'] == 'Laboratory-confirmed case']
df_confirmed_sex = df_lab_confirmed[['current_status','sex']].groupby('sex').count()
df_confirmed_sex
```

```python
explode = (0,0,0,1)
df_confirmed_sex.groupby(['sex']).sum().plot(kind='pie', y='current_status',
figsize=(10,5),explode = explode,autopct='%1.1f%%')
plt.title('confirmed cases by sex')
plt.legend(loc = 1)
```

```python
#death by sex
df_death_sex = df_lab_confirmed.loc[df_lab_confirmed['death_yn'] == 'Yes']
df_death_sex = df_death_sex[['current_status','sex']].groupby('sex').count()
df_death_sex
```

```python
explode = (0,0,0,1)
df_death_sex.groupby(['sex']).sum().plot(kind='pie', y='current_status',
figsize=(10,5), explode = explode, autopct='%1.1f%%')
plt.title('death by sex')
plt.legend(loc = 1)
```

```python
#confirmed cases by states
df_lab_confirmed = df[df['current_status'] == 'Laboratory-confirmed case']
df_comfired_state = df_lab_confirmed[['current_status','res_state']].groupby('res_state').count()
df_comfired_state
```

```python
df_confirmed_state = df_lab_confirmed[['current_status','res_state']].groupby('res_state').count()
names = df_confirmed_state.index
values = df_confirmed_state['current_status']

x = (range(56))
new_x = [.5*i for i in x]                     
x = np.arange(len(names))
fig, ax = plt.subplots(figsize=(15,5))


w = 0.4
plt.xticks(new_x , names, rotation='vertical')
ax.bar(new_x, values, width=w, color='rgbkymc', align='center')
plt.title('confirmed cases by state')
plt.ylabel('confirmed cases')


plt.show()
```

```python
df_confirmed_county = df_lab_confirmed[['current_status','res_county']].groupby('res_county').count()
names = df_confirmed_county.index
values = df_confirmed_county['current_status']


x = (range(len(names)))
new_x = [.5*i for i in x]                     
x = np.arange(len(names))
fig, ax = plt.subplots(figsize=(50,10))

w = 0.4
plt.xticks(new_x ,names,  rotation='vertical')
ax.bar(new_x, values, width=w, color='b', align='center')

plt.title('confirmed cases by county')
plt.ylabel('confirmed cases')


plt.show()
```

```python
df_confirmed_county = df_lab_confirmed[['current_status','res_county']].groupby('res_county').count()
names = df_confirmed_county.index
values = df_confirmed_county['current_status']


x = (range(len(names)))
new_x = [.5*i for i in x]                     
x = np.arange(len(names))
fig, ax = plt.subplots(figsize=(100,800))

plt.barh(new_x, values)
plt.yticks(new_x, names)

plt.title('confirmed cases by county')
plt.ylabel('confirmed cases')


plt.show()
```
